```python
#unsorted sort time(intgers)
import time
import random

#The Algorithm
def gnomeSort( arr, n):
    index = 0
    while index < n:
        if index == 0:
            index = index + 1
        if arr[index] >= arr[index - 1]:
            index = index + 1
        else:
            arr[index], arr[index-1] = arr[index-1], arr[index]
            index = index - 1
  
    return arr
  
#makes a list of random integers from -100 to 100
arr = []
for i in range(0,8):
    x = random.randint(-100, 100)
    arr.append(x)
print(arr)
n = len(arr)

#Runs Algorithm and times execution
s = time.time()
arr = gnomeSort(arr, n)
e = time.time()
print ("Sorted sequence after applying Gnome Sort :")
print(arr)
print(e-s)
```
